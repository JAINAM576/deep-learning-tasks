### 3. Legal Document Summarization
- In the context of Legal Document Summarization, PEGASUS (Pre-training with Extracted Gap-sentences for Abstractive Summarization) is a state-of-the-art model developed by Google Research.
- Domain Adaptation: While the base model is powerful, it is often fine-tuned on legal datasets (like Multi-LexSum or BillSum) to better understand specialized legal terminology (like vernietigingsbevoegdheid).

| Model   | Strength in Legal Context                                                                                       |
|---------|---------------------------------------------------------------------------------------------------------------|
| BART    | Excellent at maintaining logical flow and "reasoning" in long documents.                                      |
| T5      | Very versatile; handles "translation" tasks (e.g., Legalese to Plain English) well.                           |
| PEGASUS | Generally considered the gold standard for abstractive summarization because its pre-training objective is specifically designed for that task. |
