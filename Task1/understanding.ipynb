### Summary: TF-IDF vs Sentence-BERT

TF-IDF and Sentence-BERT both convert text to vectors, but in very different ways:
- **TF-IDF** uses word counts (sparse, large, no semantic meaning)
- **Sentence-BERT** uses neural networks to capture sentence meaning (dense, smaller, semantic)

**Example:**

- Sentence: `The hospital reform law was approved`
- **TF-IDF vector:** `[0.6, 0.7, 0.5, 0.4, 0]` (each number = a word in the vocabulary)
- **Sentence-BERT vector:** `[0.12, -0.83, 0.44, ...]` (captures meaning, not just words)

**Key Points:**

- **TF-IDF:**
  - Word-based, sparse, fast
  - No semantic understanding
  - Great for simple classification
  - Example: "hospital reform law" vs "medical system legislation" → very different vectors

- **Sentence-BERT:**
  - Meaning-based, dense
  - Captures semantics
  - Best for semantic search and similarity
  - Example: "hospital reform law" vs "medical system legislation" → very similar vectors

- For many NLP tasks, **TF-IDF + linear models** are surprisingly competitive with deep learning models.
