```python
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')

stop_words = set(stopwords.words('english'))
custom_stopwords = {"said", "could", "would", "might", "also", "one"}
stop_words.update(custom_stopwords)

tokenized_sentences = [
    [w for w in sent if w not in stop_words]
    for sent in tokenized_sentences
]

# Remove any sentences that became empty after filtering
tokenized_sentences = [sent for sent in tokenized_sentences if sent]

print(tokenized_sentences[:3])
```

## VOCAB BUILDING

```python
from collections import Counter

all_words = [word for sent in tokenized_sentences for word in sent]

vocab = Counter(all_words)

print(f"Vocab size: {len(vocab)}")
print(vocab.most_common(10))
```

```python
digits_in_vocab = [word for word in vocab if any(c.isdigit() for c in word)]
print(digits_in_vocab)
```

```python
word2idx = {word: idx for idx, word in enumerate(vocab)}
idx2word = {idx: word for word, idx in word2idx.items()}
```

```python
print(len(word2idx))
print(len(idx2word))
```

```python
word2idx = {word: idx+1 for idx, (word, _) in enumerate(vocab.most_common())}
word2idx["<PAD>"] = 0
word2idx["<UNK>"] = len(word2idx)

idx2word = {idx: word for word, idx in word2idx.items()}
```

```python
encoded_sentences = []

for sentence in tokenized_sentences:
    encoded = [word2idx.get(word, word2idx["<UNK>"]) for word in sentence]
    encoded_sentences.append(encoded)

print(encoded_sentences[:5])
```

```python

vocab_size = len(word2idx)
```

```python
from collections import Counter
word_freq = Counter([word for sent in tokenized_sentences for word in sent])
total = sum(word_freq.values())
t = 1e-5  # threshold

keep_prob = {
    word: min(1, np.sqrt(t / (word_freq[word]/total)))
    for word in word_freq
}
```

```python
import random

max_window = 5

pairs = []

for sentence in encoded_sentences:
    for i, target in enumerate(sentence):
        if random.random() > keep_prob.get(idx2word.get(target, ""), 1):
            continue  
        window_size = random.randint(1, max_window)

        start = max(0, i - window_size)
        end = min(len(sentence), i + window_size + 1)

        for j in range(start, end):
            if i != j:
                pairs.append((target, sentence[j]))


print(f"Total pairs: {len(pairs)}")
print(pairs[:10])
```

## NEGATIVE SAMLING

```python
positive_pairs = [(target, context, 1) for target, context in pairs]

negative_pairs = []

for target, context, _ in positive_pairs:
    for _ in range(3): 

        while True:
            neg_context = random.randint(1, vocab_size - 1)
            if neg_context != context:
                break

        negative_pairs.append((target, neg_context, 0))

all_pairs = positive_pairs + negative_pairs

all_pairs = np.array(all_pairs)
np.random.shuffle(all_pairs)

X_target = all_pairs[:, 0].astype(np.int32)
X_context = all_pairs[:, 1].astype(np.int32)
y = all_pairs[:, 2].astype(np.float32)

print(f"Positive pairs: {len(positive_pairs)}")
print(f"Negative pairs: {len(negative_pairs)}")
print(f"Total: {len(all_pairs)}")
```

## MODEL TRAINING

## NEGATIVE SAMPLING TRAINING

```python
# If (target, context) is real → make embeddings similar
# If fake → push embeddings apart
```

```python
# uniform negative sampling
# dot → Dense → sigmoid
# 5 negatives per positive
```

```python
import tensorflow as tf

vocab_size = len(word2idx) + 1

input_target = tf.keras.Input(shape=(1,))
input_context = tf.keras.Input(shape=(1,))

emb = tf.keras.layers.Embedding(vocab_size, 300)

target_emb = emb(input_target)
context_emb = emb(input_context)

dot = tf.keras.layers.Dot(axes=2)([target_emb, context_emb])
dot = tf.keras.layers.Reshape((1,))(dot)
output = tf.keras.layers.Dense(1, activation='sigmoid')(dot)

model = tf.keras.Model([input_target, input_context], output)
model.compile(loss='binary_crossentropy', optimizer='adam')
model.summary()
```

```python


model.fit(
    [X_target, X_context],
    y,
    epochs=5,
    batch_size=4096
)
```

```python
embeddings = model.get_layer('embedding_17').get_weights()[0]
print(embeddings.shape)
```

```python

def get_embedding(word):
    idx = word2idx.get(word, word2idx["<UNK>"])
    return embeddings[idx]

print(get_embedding("harry").shape) 
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def most_similar(word, top_n=10):
    idx = word2idx.get(word, word2idx["<UNK>"])
    vec = embeddings[idx].reshape(1, -1)
    sims = cosine_similarity(vec, embeddings)[0]
    top_idx = np.argsort(sims)[::-1][1:top_n+1]
    return [(idx2word[i], round(sims[i], 4)) for i in top_idx]

print(most_similar("harry"))
print(most_similar("hermione"))
print(most_similar("dumbledore"))
print(most_similar("father"))
```
