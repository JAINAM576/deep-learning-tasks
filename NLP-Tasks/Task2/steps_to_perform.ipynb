## Standard NLP Preprocessing Pipeline

1. Lowercasing
- "Harry", "harry", "HARRY" → same word
- Avoids duplicate tokens

2. Remove Noise (Punctuation, Special Characters)
- Remove symbols like .,!?@#
- Removes meaningless tokens
- Cleaner input → better embeddings

3. Tokenization
- "Harry met Ron. Hermione smiled."
→ ["Harry met Ron", "Hermione smiled"]

4. Remove Stopwords (Optional)
- ["the", "is", "at", "on"]
- Removes low-information words
- Improves focus on meaningful terms

5. Stemming
- Reduce words to root form (crude)
- running → run
- studies → studi
- Reduces vocabulary
- Groups similar words

6. Lemmatization
- Convert to dictionary form
- running → run
- better → good
- More accurate than stemming
- Keeps meaning intact

7. Remove Rare Words
- Remove words that appear very few times
- "azkaban" appears 1 time → removed

- Reduces noise
- Faster training

8. Remove Frequent Words
- Remove extremely common words
- "said", "mr", "one"
- Improves embedding quality

9. Vocabulary Creation
- Map words → numbers
- harry → 0
- ron → 1
- hermione → 2
- Models only understand numbers
- Required for embedding layers

10. Padding / Truncation
- Make sequences same length
- ["harry", "met"] → [harry, met, PAD]
- Required for batch training

11. Vectorization
- Convert words → vectors
- Methods:
-TF-IDF
-Word2Vec
- Embeddings (your case)

- Enables mathematical operations
- Captures meaning

## Order
1. Text Collection
2. Lowercasing
3. Remove Noise (punctuation, special chars, numbers)
4. Sentence Tokenization
5. Word Tokenization
6. Stopword Removal (optional)
7. Stemming / Lemmatization (optional)
8. Remove Rare / Frequent Words (optional)
9. Vocabulary Creation (word → index)
10. Convert text → sequences (numbers)
11. Padding / Truncation (if needed)
12. Vectorization / Embedding
